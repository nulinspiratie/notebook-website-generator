# Turnon <a id='Turnon'></a>

In the first measurement we perform a turnon scan, simultaneously sweeping TG, LB, and RB

```python
voltages = np.linspace(0, 2, 201)
currents = 1 / (1 + np.exp(15*(-voltages+1)))
plt.figure()
plt.plot(voltages, currents)
plt.xlabel('Top gate, RB, LB (V)')
plt.ylabel('Current (nA)')
```

# Pinchoff

Next we perform pinchoff, we can see that both gates pinch off at ~0.5V, RB a bit higher than LB

```python
voltages = np.linspace(0, 2, 201)
plt.figure()
currents = 1 / (1 + np.exp(15*(-voltages+0.5)))
plt.plot(voltages, currents, label='LB')
currents = 1 / (1 + np.exp(15*(-voltages+0.55)))
plt.plot(voltages, currents, label='RB')
plt.xlabel('Top gate, RB, LB (V)')
plt.ylabel('Current (nA)')
plt.legend()
```
